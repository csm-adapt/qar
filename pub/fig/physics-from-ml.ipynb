```python
import numpy as np
import matplotlib as mpl
import seaborn as sns
from matplotlib import cm
from matplotlib import pyplot as plt
from scipy.special import factorial
```

```python
def exp_approx(x, order=1):
  k = np.arange(order+1)
  return np.sum(np.power.outer(x, k)/factorial(k), axis=1)

x = np.linspace(-1, 3, num=100)

y = np.exp(x)

y1 = exp_approx(x, order=1)
y2 = exp_approx(x, order=2)
y3 = exp_approx(x, order=3)
y4 = exp_approx(x, order=4)
y5 = exp_approx(x, order=5)
y6 = exp_approx(x, order=6)
y7 = exp_approx(x, order=7)
y8 = exp_approx(x, order=8)
y9 = exp_approx(x, order=9)

font = {
  'family': 'serif',
  'size': 28
}
mpl.rc('font', **font)
mpl.rc('text', **{'usetex': True})

fig, ax = plt.subplots(figsize=(16,9));

plot = ax.plot

plot(x, y, 'k', lw=3);
#
plot(x, y1, linestyle=':', color="violet", lw=3, label="N=1");
plot(x, y2, linestyle='--', color="mediumorchid", lw=3, label="N=2");
plot(x, y3, linestyle='-', color="darkviolet", lw=3, label="N=3");
#
# plot(x, y4, linestyle=':', color="springgreen", lw=3, label="N=4");
# plot(x, y5, linestyle='--', color="mediumseagreen", lw=3, label="N=5");
# plot(x, y6, linestyle='-', color="seagreen", lw=3, label="N=6");
plot(x, y4, linestyle=':', color="gold", lw=3, label="N=4");
plot(x, y5, linestyle='--', color="orange", lw=3, label="N=5");
plot(x, y6, linestyle='-', color="orangered", lw=3, label="N=6");
#
# plot(x, y7, linestyle=':', color="gold", lw=3, label="N=7");
# plot(x, y8, linestyle='--', color="orange", lw=3, label="N=8");
# plot(x, y9, linestyle='-', color="darkorange", lw=3, label="N=9");

plt.legend(frameon=False, title=r'$e^x \sim \sum_{k=0}^N \frac{x^k}{k!}$');

ax.set_ylim(bottom=0.3);
ax.spines['top'].set_visible(False);
ax.spines['right'].set_visible(False);
ax.spines['bottom'].set_linewidth(3);
ax.spines['left'].set_linewidth(3);

ax.xaxis.set_tick_params(width=3);
ax.yaxis.set_tick_params(width=3);

ax.set_xlabel("$x$", fontsize=32);
ax.set_ylabel(r"$e^x$", fontsize=32);

ax.set_ylim(bottom=0);

sns.despine(ax=ax, offset=10, trim=True);

fig.savefig("polynomial-approx.pdf", bbox_inches="tight");
```
